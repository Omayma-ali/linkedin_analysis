### üóëÔ∏è Dropping Unimportant Columns

```python
# delete unimportant columns
df = df.drop(['Unnamed: 0', 'posLocationCode', 'posLocation', 'endDate', 'companyHasLogo', 'companyUrn', 'companyUrl', 'mbrLocationCode', 'memberUrn', 'positionId'], axis=1)
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
# converting 'haspicture' column to 0 and 1 values
df['hasPicture'].unique()
```

```python
# if the profile has picture then put 1 else put 0
df['hasPicture'] = df['hasPicture'].apply(lambda x : 1 if pd.notna(x) else 0)
```

```python
df['hasPicture'].value_counts()
```

```python
df.isna().sum()
```

```python
# drop missing values for the remaining column
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

```python
# shape after drop the missing values
df.shape
```

```python
# Correct data types for columns
df['companyFollowerCount'] = df['companyFollowerCount'].astype(int)
df['companyStaffCount'] = df['companyStaffCount'].astype(int)
df['avgMemberPosDuration'] = df['avgMemberPosDuration'].astype(int)
df['avgCompanyPosDuration'] = df['avgCompanyPosDuration'].astype(int)
```

# 5- üìä Data Visualization 

```python
# Correlation Matrix
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, fmt='.2f', cmap='coolwarm', square=True, cbar_kws={"shrink": .8})
plt.title('Correlation Matrix')
plt.show()
```

```python
# Pairplot
plt.figure(figsize=(12, 20))
sns.pairplot(df)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='genderEstimate', data=df)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
gender_distribution = df['genderEstimate'].value_counts(normalize=True) * 100
print(f'Gender distribution: {gender_distribution.to_dict()}')
```

```python
# positions frequency
position_counts = df['posTitle'].value_counts().head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=position_counts.index, y=position_counts.values)
plt.title('Top 10 Job Titles')
plt.xlabel('Job Title')
plt.ylabel('Number of Members')
plt.xticks(rotation=45)
plt.show()
```

```python
# Most popular companies
top_companies = df['companyName'].value_counts().head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_companies.index, y=top_companies.values)
plt.title('Top 10 Companies by Number of Members')
plt.xlabel('Company Name')
plt.ylabel('Number of Members')
plt.xticks(rotation=45)
plt.show()
```

```python
# Follower Age Frequency
plt.figure(figsize=(10, 6))
sns.histplot(df['ageEstimate'], kde=True)
plt.title('Distribution of Followers Age')
plt.xlabel('Follower Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Age analysis
avg_age = df['ageEstimate'].mean()
print(f'The average age of members is approximately {avg_age:.2f} years.')
```

```python
top_companies = df.groupby('companyName')['companyStaffCount'].max().sort_values(ascending=False).head(10).reset_index()
# Number of Staff in Top 10 Companies
plt.figure(figsize=(10, 6))
sns.barplot(x='companyName', y='companyStaffCount', data=top_companies, palette='viridis')
plt.title('Number of Staff in Top 10 Companies')
plt.xlabel('Company Name')
plt.ylabel('Number of Staff')
plt.xticks(rotation=45, ha='right')  
plt.tight_layout()  
plt.show()
```

```python
largest_company_count = df['companyStaffCount'].max()
largest_company = df.loc[df['companyStaffCount'].idxmax(), 'companyName']
print(f'The largest company in the dataset is {largest_company} with {largest_company_count} employees.')
```

```python
premium_members = df[df['isPremium'] == 1].shape[0]
total_members = df.shape[0]
non_premium_members = total_members - premium_members
premium_percentage = (premium_members / total_members) * 100

# Percentage of Premium Members
plt.figure(figsize=(6, 6))
labels = ['Premium Members', 'Non-Premium Members']
sizes = [premium_members, non_premium_members]
colors = ['#4CAF50', '#FFC107']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, explode=[0.1, 0])
plt.title('Percentage of Premium Members')
plt.show()
```

```python
average_followers =  df.groupby('companyName')['followersCount'].mean().sort_values(ascending=False).head(10).reset_index()

# Average Number of Followers foreach company
plt.figure(figsize=(10, 6))
sns.barplot(x='companyName', y='followersCount',data=average_followers, palette='Blues_r')
plt.title('Average Number of Followers per company')
plt.xlabel('Company Name')
plt.ylabel('followersCount')
plt.xticks(rotation=45, ha='right')  
plt.tight_layout()  
plt.show()
```

```python
# Relationship between Connections and Followers
plt.figure(figsize=(8, 6))
sns.scatterplot(x='connectionsCount', y='followersCount', data=df, color='#FF5733', alpha=0.6)
sns.regplot(x='connectionsCount', y='followersCount', data=df, scatter=False, color='blue', line_kws={'linewidth': 2})
plt.title(f'Relationship between Connections and Followers')
plt.xlabel('Connections Count')
plt.ylabel('Followers Count')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
df['startDate'] = pd.to_datetime(df['startDate'])
df['year'] = df['startDate'].dt.year
yearly_counts = df['year'].value_counts().sort_index()
yearly_counts.plot(kind='bar', color='darkorange')
plt.title('Number of Positions Started Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Positions')
plt.show()
```

```python
most_active_year = yearly_counts.idxmax()
most_active_year_count = yearly_counts.max()
print(f'The most active year for new positions was {most_active_year} with {most_active_year_count} positions started.')
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(df['avgMemberPosDuration'], label='Avg Member Position Duration', color='red')
sns.kdeplot(df['avgCompanyPosDuration'], label='Avg Company Position Duration', color='blue')
plt.title('Distribution of Position Durations (Member vs. Company)')
plt.legend()
plt.show()
```

```python
# Position duration analysis
avg_member_duration = df['avgMemberPosDuration'].mean()
avg_company_duration = df['avgCompanyPosDuration'].mean()
print(f'The average position duration for members is {avg_member_duration:.2f} months,\nwhile the average position duration for companies is {avg_company_duration:.2f} months.')
```

```python
top_10_companies = df['companyName'].value_counts().head(10).index
df_top_companies = df[df['companyName'].isin(top_10_companies)]

# Age vs. CompanyName (Scatter Plot)
plt.figure(figsize=(10, 6))  
sns.scatterplot(y='ageEstimate', x='companyName', hue='genderEstimate', data=df_top_companies, palette='coolwarm', s=100, alpha=0.7)
plt.title('Age vs. Company Name (Colored by Gender)')
plt.xlabel('Company Name')
plt.ylabel('Age Estimate')
plt.xticks(rotation=45, ha='right')  
plt.tight_layout()
plt.show()
```
